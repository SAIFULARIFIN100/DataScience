6.Create a new column called CustomerValue with criteria: if TotalCharges >= median then 'High Value' else 'Low Value’.

```python
TelcoCustomer['TotalCharges'] = TelcoCustomer['TotalCharges'].replace(to_replace = ' ', value = np.NaN)
TelcoCustomer[TelcoCustomer['TotalCharges'].isna()]
```

```python
TelcoCustomer['TotalCharges']=TelcoCustomer['TotalCharges'].astype(float)
```

```python
TelcoCustomer['median_TotalCharges']=TelcoCustomer['TotalCharges'].median()
```

```python
def TotalCharges_cat (x):
  if x >= TelcoCustomer['TotalCharges'].median():
    return "High Value"
  elif x < TelcoCustomer['TotalCharges'].median():
     return "Low Value"
```

```python
TelcoCustomer['CustomerValue']=TelcoCustomer['TotalCharges'].apply(TotalCharges_cat)
TelcoCustomer.head()
```

10.Encode Contract column (assume it has hierarchy).

```python
# Label encoding for Contract
def Contract_yes(x):
    if x == 'Month-to-month':
        return 3
    elif x == 'One year':
        return 2
    elif x == 'Two year':
        return 1
  
    
TelcoCustomer['Contract'] = TelcoCustomer['Contract'].apply(Contract_yes)
```
