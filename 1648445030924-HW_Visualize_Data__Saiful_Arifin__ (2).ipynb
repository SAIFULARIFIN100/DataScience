```python
#import matplotlib dan seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
import numpy as np
#Your codes here
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
labels = ['Yes','No']
churn = df.Churn.value_counts()
ax.pie(churn, labels=labels, autopct='%.0f%%')
plt.show()
```

```python
df_churn_total = df['Churn'].value_counts(normalize=True).rename_axis('Churn').reset_index(name='Customer_Percentage')
df_churn_total['Customer_Percentage'] = df_churn_total['Customer_Percentage']*100
df_churn_total
```

```python
df_churn_total = df['Churn'].value_counts(normalize=False).rename_axis('Churn').reset_index(name='Total_User')
df_churn_total['Total_User'] = df_churn_total['Total_User']*100
df_churn_total
```

```python
sns.barplot(x = 'Churn', y = 'Customer_Percentage', data=df_churn_total)
```

```python
sns.barplot(x = 'Churn', y='Total_User', data=df_churn_total)
```

```python
df["TotalCharges"].value_counts()
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(to_replace = ' ', value = np.NaN)
df[df['TotalCharges'].isna()]
```

```python
df['TotalCharges']=df['TotalCharges'].astype(float)
```

```python
sns.displot(data=df, x="TotalCharges", hue="Churn", kind="kde", ax=ax)
```

```python
def pairplot(dataset, target_column):
    sns.set(style="ticks")
    sns.pairplot(dataset, hue=target_column)
    plt.show()
```

```python
pairplot(df, "Churn")
```
